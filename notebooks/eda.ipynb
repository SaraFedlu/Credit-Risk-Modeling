```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../data/data.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
# Distribution of Numerical Features
numerical_features = ['Amount', 'Value', 'FraudResult']
for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[feature], kde=True, bins=30, color='blue')
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Distribution of Categorical Features
categorical_features = ['CurrencyCode', 'CountryCode', 'ProviderId', 'ProductCategory', 'ChannelId']
for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=data, x=feature, palette='viridis', order=data[feature].value_counts().index)
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Correlation Matrix
correlation_matrix = data[numerical_features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Identifying Outliers
for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=data[feature], palette='Set2')
    plt.title(f"Boxplot for {feature}")
    plt.xlabel(feature)
    plt.show()
```

```python
# Relationship with FraudResult
for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='FraudResult', y=feature, data=data, palette='coolwarm')
    plt.title(f"{feature} by FraudResult")
    plt.xlabel('FraudResult')
    plt.ylabel(feature)
    plt.show()
```

```python
# Drop Irrelevant Columns
data_cleaned = data.drop(columns=['CurrencyCode', 'CountryCode'])
```

```python
# Handling Outliers
def handle_outliers_log_transform(df, columns):
    for col in columns:
        # Add 1 to avoid log(0)
        df[col + '_log'] = np.log1p(df[col])
    return df

# Apply log transformation to Amount and Value
numerical_features = ['Amount', 'Value']
data_cleaned = handle_outliers_log_transform(data_cleaned, numerical_features)
```

```python
# Group infrequent ProviderId categories
provider_threshold = 500
data_cleaned['ProviderId'] = data_cleaned['ProviderId'].apply(
    lambda x: x if data_cleaned['ProviderId'].value_counts()[x] >= provider_threshold else 'Other'
)

# Group infrequent ProductCategory categories
category_threshold = 500
data_cleaned['ProductCategory'] = data_cleaned['ProductCategory'].apply(
    lambda x: x if data_cleaned['ProductCategory'].value_counts()[x] >= category_threshold else 'Other'
)
```
